```python
# import libraries
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# notebook to graph individual plots

# use xarray to open dataset
file_of_interest = ""
ds = xr.open_dataset(file_of_interest)
data = ds['Band1']
ds.close()

# index every other to help with memory
data = data[::2, ::2] 
masked_data = np.ma.masked_where(data == np.nan, data)

# Create coordinate arrays for lat and lon
x = ds['lon'][::2]
y = ds['lat'][::2]

# graph using pcolormesh
fig, ax = plt.subplots()
im = ax.pcolormesh(x, y, masked_data, cmap='viridis', shading='auto',vmin = -40, vmax = 40)
ax.set_aspect('equal')
cbar2 = fig.colorbar(im)
cbar2.set_label('Topography/Bathymetry', fontsize = 11)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
plt.show()
```

```python
# notebook to graph file and its surrounding ones. Make sure to include the file_of_interest as the 5 file in surrounding_file_names for indexing purposes


file_of_interest = ""
# enter surrounding names from excel table created in full_program
surrounding_file_names = ["", "", "", "", file_of_interest, "", "", "", ""]
    #N, S, E, W, NW, NE, SW, SE


index = 0
fig, axes = plt.subplots(3, 3, figsize=(9, 9))
axes = axes.flatten()

# assign north, east, south, west locations for graphing boxes in 3 x 3
for i, item in enumerate(surrounding_file_names):
    if i == 0:
        index = 1
    elif i == 1:
        index = 5
    elif i == 2:
        index = 7
    elif i == 3:
        index = 3
    elif i == 4:
        index = 4
    elif i == 5:
        index = 0
    elif i == 6:
        index = 2
    elif i == 7:
        index = 6
    elif i == 8:
        index = 8
    else:
        print("error")


    # get data with same method as individual graph
    ds = xr.open_dataset(item)
    var_data = ds['Band1'].values
    var_data = var_data[::2, ::2]
    Y = ds['lat'][::2]
    X = ds['lon'][::2]
    im = axes[index].pcolormesh(X, Y, var_data, cmap='viridis', shading = 'auto', vmin = -40, vmax = 40)
    axes[index].set_aspect('equal')


# graph on subplots
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
cbar = fig.colorbar(im, cax=cbar_ax)
fig.supxlabel('Longitude', fontsize=14)
fig.supylabel('Latitude', fontsize=14)
cbar.set_label('Topography/Bathymetry', fontsize = 12)

plt.tight_layout(rect=[0, 0, 0.9, 1])  # Leave space for colorbar
```
